## Задача 5

```python
import seaborn as sea
import matplotlib.pyplot as plt
import numpy as np
```

Гистограмма распределения значения редкости персонажей

```python
figure = plt.figure(figsize=(4, 4))
axis = figure.add_subplot(111)
numberOfUniqueValues = data.rarity.drop_duplicates().count()
axis.set_xlim((data.rarity.min()), data.rarity.max())
axis.hist(data.rarity, bins=numberOfUniqueValues, color="#5674ce")
plt.savefig("images/rarity_histogram.png")
```

Диаграмма "ящик с усами" со значениями атаки персонажей

```python
figure = plt.figure(figsize=(4, 4))
axis = figure.add_subplot(111)
sea.boxplot(x=data.DPS, ax=axis, color="#56b8ce", flierprops={"marker": "*"})
plt.savefig("images/attack_boxplot.png")
```

Круговая диаграмма классов персонажей

```python
figure = plt.figure(figsize=(4, 4))
axis = figure.add_subplot(111)
rolesCount = data.operator_role.value_counts()
rolesLabels = rolesCount.index
axis.pie(rolesCount, labels=rolesLabels, autopct="%1.1f%%", colors=["#5cc9a8", "#e2a46a", "#9462e5", "#8284a5", "#d8e562", "#6bd369", "#ce69d3", "#6aa0e2"]);
plt.savefig("images/roles_pie.png")
```

Карта корреляции числовый значений

```python
figure = plt.figure(figsize=(6, 5))
axis = figure.add_subplot(111)
sea.heatmap(data[["rarity", "DPS", "redeploy_time", "dp_cost"]].corr(), annot=True, cmap="bone", ax=axis)
plt.savefig("images/values_heatmap.png")
```

Диаграмма с группировкой по классам персонажей и количеству целей во время атаки

```python
figure = plt.figure(figsize=(6, 5))
axis = figure.add_subplot(111)
sea.countplot(x=data.operator_role, hue=data.target)
plt.savefig("images/roles_to_target_countplot.png")
```
